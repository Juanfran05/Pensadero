```python
df = pd.read_csv('Acamica/ds-online-40/Notebooks/Data/DS_Proyecto_01_Datos_Properati.csv')
df.head(3)
```

```python
useless_cols = ['start_date', 'end_date', 'created_on', 'lat', 'lon', 'currency', 'title', 'description',  'operation_type']
df.drop(columns=useless_cols, inplace=True)
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
columns = ['rooms','bedrooms','bathrooms','surface_total','surface_covered','price']
```

## StandarScaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X = scaler.fit_transform(df[columns])
```

```python
X = pd.DataFrame(X, columns=columns)

print('Outliers por columnas')
(X>3).sum()
```

```python
df[X.bathrooms>3].bathrooms.unique()
```

#### Propiedades con multi-variante outliers

```python
df.loc[(((X>3).sum(axis=1) / len(columns)) > 0.5)].head(15)
```

## DBSCAN

```python
from sklearn.preprocessing import RobustScaler
from sklearn.cluster import DBSCAN
```

```python
robus_scaler = RobustScaler(quantile_range=(0.1, 0.9))
X = robus_scaler.fit_transform(df[columns])
```

```python
dbscan = DBSCAN(eps=5, min_samples=3)
```

```python
dbscan.fit(X)
```

```python
X = pd.DataFrame(X, columns=columns)
X.loc[:, 'labels'] = dbscan.labels_
```

```python
X[X.labels == -1]
```

## IsolationForest

```python
from sklearn.ensemble import IsolationForest
```

```python
iso = IsolationForest(n_jobs=-1)
X = iso.fit_predict(df[columns])
```
